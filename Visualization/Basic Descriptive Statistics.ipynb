```python
# 数据正态性检验（shapiro-Wilk）检验
datas = [df_filter1['点赞数'].tolist(),df_filter2['评论数'].tolist(),df_filter3['分享数'].tolist(),df_filter4['收藏数'].tolist()]
for data in datas:
    shapiro_stat, shapiro_pvalue = shapiro(data)
    print(f"Shapiro-Wilk Statistic: {shapiro_stat}")
    print(f"P-value: {shapiro_pvalue}")

    if shapiro_pvalue < 0.05:
        print("数据不服从正态分布")
    else:
        print("数据服从正态分布")
```

```python
## 对服从正太分布的用ols回归

# 需要转换的列
columns_to_convert = ['#数量', '人脸数量','视频时长', 'length_dlut', 'positive_dlut', 'negative_dlut']

# 先填充缺失值为0，再将这些列转换为 float 类型
df_filter1[columns_to_convert] = df_filter1[columns_to_convert].fillna(0).astype(float)

X = df_filter1[[ '#数量','人脸数量','视频时长','tempo', 'mean_rms','avg_hue',
                'avg_brightness', 'avg_saturation','length_dlut', 'positive_dlut', 'negative_dlut']]

y = df_filter1['点赞数']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
# model.summary()

# 获取回归结果的摘要
summary = model.summary2().tables[1]

# 添加星号标记
def significance_stars(p):
    if p < 0.001:
        return '***'
    elif p < 0.01:
        return '**'
    elif p < 0.05:
        return '*'
    else:
        return ''

summary['P>|t|'] = summary['P>|t|'].astype(float)
summary['Significance'] = summary['P>|t|'].apply(significance_stars)

summary
```

```python
## 对不属于正太分布的跑个泊松分布

# 需要转换的列
columns_to_convert = ['#数量', '人脸数量','视频时长', 'length_dlut', 'positive_dlut', 'negative_dlut']

# 先填充缺失值为0，再将这些列转换为 float 类型
df_filter3[columns_to_convert] = df_filter3[columns_to_convert].fillna(0).astype(float)

# 特征变量
X = df_filter3[[ '#数量','人脸数量','视频时长','tempo', 'mean_rms','avg_hue',
                'avg_brightness', 'avg_saturation','length_dlut', 'positive_dlut', 'negative_dlut']]

# 目标变量
y = df_filter3['分享数']

# 添加常数项
X = sm.add_constant(X)

# 拟合泊松回归模型
model = sm.GLM(y, X, family=sm.families.Poisson()).fit()

# 获取回归结果的摘要
summary = model.summary2().tables[1]

# 添加星号标记
def significance_stars(p):
    if p < 0.001:
        return '***'
    elif p < 0.01:
        return '**'
    elif p < 0.05:
        return '*'
    else:
        return ''

summary['P>|z|'] = summary['P>|z|'].astype(float)
summary['Significance'] = summary['P>|z|'].apply(significance_stars)

# 打印摘要
summary
```

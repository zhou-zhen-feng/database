##  情感描述性统计，1850—1875.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
```

### 以月为单位

```python
Positive,Negative,Anger,Anticipation,Disgust,Fear,Joy,Sadness,Surprise,Trust = [[] for _ in range(10)]
Yearmonths = []
folder_path = r"D:\American Stories\the civil war\emotion_score"
for file_name in tqdm(os.listdir(folder_path)):
    file_path = os.path.join(folder_path,file_name)
    df = pd.read_csv(file_path)
    positive = df["positive"] / df["length"]
    negative = df["negative"] / df["length"]
    anger = df["anger"] / df["length"]
    anticipation = df["anticipation"] / df["length"]
    disgust = df["disgust"] / df["length"]
    fear = df["fear"] / df["length"]
    joy = df["joy"] / df["length"]
    sadness = df["sadness"] / df["length"]
    surprise = df["surprise"] / df["length"]
    trust = df["trust"] / df["length"]
    
    Positive.append(positive.mean())
    Negative.append(negative.mean())
    Anger.append(anger.mean())
    Anticipation.append(anticipation.mean())
    Disgust.append(disgust.mean())
    Fear.append(fear.mean())
    Joy.append(joy.mean())
    Sadness.append(sadness.mean())
    Surprise.append(surprise.mean())
    Trust.append(trust.mean())
    
    yearmonth = file_name.split('.')[0]
    Yearmonths.append(yearmonth)
```

#### 1850-1875年，全部情感类别可视化

```python
plt.figure(figsize=(12, 8))

# 绘制每个情绪的折线图
plt.plot(Yearmonths, Anger, label='Anger', color='red', linestyle='-')
plt.plot(Yearmonths, Anticipation, label='Anticipation', color='orange', linestyle='-')
plt.plot(Yearmonths, Disgust, label='Disgust', color='green', linestyle='-')
plt.plot(Yearmonths, Fear, label='Fear', color='blue', linestyle='-')
plt.plot(Yearmonths, Joy, label='Joy', color='purple', linestyle='-')
plt.plot(Yearmonths, Sadness, label='Sadness', color='brown', linestyle='-')
plt.plot(Yearmonths, Surprise, label='Surprise', color='pink', linestyle='-')

# 设置图例
plt.legend(loc='center left', fontsize='medium', bbox_to_anchor=(1, 0.5))

# 设置 x 轴标签的角度和显示
plt.xticks(Yearmonths[::24], rotation=45)  # 每隔12个月显示一个刻度，45度旋转标签

# 调整y轴范围
plt.ylim(0.005, 0.025)

# 添加标题和标签
plt.title('Emotion Trends Over Time', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Value', fontsize=14)

# 启用网格
plt.grid(True, linestyle='--', alpha=0.7)

# 显示图表
plt.tight_layout()  # 自动调整子图参数，以填充整个图像区域
plt.show()
```

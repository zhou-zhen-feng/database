#  描述性统计，各个道德化维度随时间的变化（1850—1875）

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
```

## 以月度为单位可视化

```python
# 计算每一年_月的道德的数据。
# care，fairness，loyalty，authority，purity

folder_path = r"D:\American Stories\the civil war\moral_score\about_black_article"
dict_moral = {}

for file_name in os.listdir(folder_path):
    file_path = os.path.join(folder_path,file_name)
    df = pd.read_csv(file_path)
    mean_values = df.mean(axis=0)
    # print(mean_values)
    mean_values = mean_values.tolist()
    # print( mean_values)
    #print(mean_values)
    year_month = file_name.split('.')[0]
    # print(year)
    dict_moral[year_month] =  mean_values 
```

```python
# dict_moral
```

```python
care = []
fairness = []
loyalty = []
authority = []
purity = []
years =[]
for key,value in tqdm(dict_moral.items()):
    care.append(value[0])
    fairness.append(value[1])
    loyalty.append(value[2])
    authority.append(value[3])
    purity.append(value[4])
    years.append(key)
```

## 5个维度放在一起

```python
plt.figure(figsize=(12, 8))

# 绘制每个维度的折线图
plt.plot(years, care, label='care', color='red', linestyle='-')
plt.plot(years, fairness, label='fairness', color='blue', linestyle='-')
plt.plot(years, loyalty, label='loyalty', color='green', linestyle='-')
plt.plot(years, authority, label='authority', color='purple', linestyle='-')
plt.plot(years, purity, label='purity', color='orange', linestyle='-')

# 设置图例
plt.legend(loc='center left', fontsize='medium', bbox_to_anchor=(1, 0.5))

# 设置 x 轴标签的角度和显示
plt.xticks(years[::24], rotation=45)  # 每隔12个月显示一个刻度，45度旋转标签

# 调整y轴范围
plt.ylim(3, 7.5)

# 添加标题和标签
plt.title('Moral Trends Over Time', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Value', fontsize=14)

# 启用网格
plt.grid(True, linestyle='--', alpha=0.7)

# 自动调整布局
plt.tight_layout()  # 自动调整子图参数，以填充整个图像区域

# 显示图表
plt.show()
```

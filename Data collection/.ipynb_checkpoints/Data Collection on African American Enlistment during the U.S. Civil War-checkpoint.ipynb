```python
import requests
import warnings
import pandas as pd
from tqdm.notebook import tqdm
from pyquery import PyQuery as pq
```

```python
def collect_urls(number):
    url = f"https://afroamcivilwar.org/soldier/page/{number}/?wpv_view_count=7633"
    response = requests.get(url)

    # 检查请求是否成功
    if response.status_code == 200:
        content = response.text
        query = pq(content)

        blacks_elem = query.find("div[class='tb-grid-column']")
        blacks_url_elem = blacks_elem.find('p a')

        urls = [a.attr('href') for a in blacks_url_elem.items() if a.attr('href')]
        # print(urls)

    else:
        print(f"请求失败，页面链接: {url}")
        urls = []
    
    return urls
```

```python
def main(path):
    for number in range(1,12460):
        urls = collect_urls(number)
        with open(path, 'a', encoding='utf-8') as file:
            for url in urls:
                file.write(url + '\n') 
```

```python
def open_urls(path):
    with open(path,'r',encoding = 'utf-8') as file:
        urls = file.read()
    url_list = urls.split('\n')
    return url_list
```

```python
def store_json(url_list,path_1):
    with open(path_1,'w') as file:
        json.dump(url_list, file, ensure_ascii=False, indent=4)
```

```python
def single_url_crawler(url):
    response = requests.get(url)
    
    # 检查请求是否成功
    if response.status_code == 200:
        content = response.text
        query = pq(content)
        
        
        soldier_name_elem = query.find("div[class='tb-fields-and-text']").children('h1')
        
        content = query.find("div.wp-block-toolset-blocks-grid-column.tb-grid-column.tb-grid-align-top[data-toolset-blocks-grid-column='3034fbe886c11054e95b46b09d3e4112']")
        spcific_content = content.find("div[class='tb-fields-and-text']").children('p')
        
        
        soldier_name = soldier_name_elem.text().split('–')[0].strip()
        #print(spcific_content)
        for item in spcific_content.items():
            #print(item)
            text = item.text()
            
            if 'NAID'in text:
                continue
                
            else:
                if 'Unit:' in text:
                    Unit = text.split('Unit:')[1].strip()
                if 'Enlisted term years:' in text:
                    Enlisted_term_years = text.split('Enlisted term years:')[1].strip()
                if 'Regiment:' in text:
                    Regiment = text.split('Regiment:')[1].strip()
                if 'ID:' in text:
                    ID = text.split('ID:')[1].strip()
                if 'Company:' in text:
                    Company = text.split('Company:')[1].strip()
                if 'Year:' in text:
                    Year = text.split('Year:')[1].strip()
                if 'Year Enlisted:' in text:
                    Year_Enlisted = text.split('Year Enlisted:')[1].strip()
                if 'Enlisted term years:' in text:
                    Enlisted_term_years = text.split('Enlisted term years:')[1].strip()
                if 'Enlisted date:' in text:
                    Enlisted_date = text.split('Enlisted date:')[1].strip()
                if 'Enlisted state:' in text:
                    Enlisted_state = text.split('Enlisted state:')[1].strip()
                if 'Enlisted county:' in text:
                    Enlisted_county = text.split('Enlisted county:')[1].strip()
                if 'Enlisted town:' in text:
                    Enlisted_town = text.split('Enlisted town:')[1].strip()
                if 'State born:' in text:
                    State_born = text.split('State born:')[1].strip()
                if 'County born:' in text:
                    County_born = text.split('County born:')[1].strip()
                if 'Town born:' in text:
                    Town_born = text.split('Town born:')[1].strip()
                if 'Occupation:' in text:
                    Occupation = text.split('Occupation:')[1].strip()
                if 'Age:' in text:
                    Age = text.split('Age:')[1].strip()
                if 'Height:' in text:
                    Height = text.split('Height:')[1].strip()
                if 'Complexion:' in text:
                    Complexion = text.split('Complexion:')[1].strip()
                if 'Eye color:' in text:
                    Eye_color = text.split('Eye color:')[1].strip()
                if 'Hair color:' in text:
                    Hair_color = text.split('Hair color:')[1].strip()
        
        info_dict = {
                'ID': ID,
                'soldier_name': soldier_name,
                'Age': Age,
                'Unit': Unit,
                'Enlisted_term_years': Enlisted_term_years,
                'Regiment': Regiment,
                'Company': Company,
                'Year': Year,
                'Year_Enlisted': Year_Enlisted,
                'Enlisted_date': Enlisted_date,
                'Enlisted_state': Enlisted_state,
                'Enlisted_county': Enlisted_county,
                'Enlisted_town': Enlisted_town,
                'State_born': State_born,
                'County_born': County_born,
                'Town_born': Town_born,
                'Occupation': Occupation,
                'Height': Height,
                'Complexion': Complexion,
                'Eye_color': Eye_color,
                'Hair_color': Hair_color
            }

        
    else:
        print(f"请求失败,链接: {url}")
        info_dict={}
    
    return info_dict
```

```python
def main_2(url_list):
    black_soldiers = pd.DataFrame(columns=[
    'ID', 'soldier_name', 'Age', 'Unit', 'Enlisted_term_years', 'Regiment', 
    'Company', 'Year', 'Year_Enlisted', 'Enlisted_date', 'Enlisted_state', 
    'Enlisted_county', 'Enlisted_town', 'State_born', 'County_born', 
    'Town_born', 'Occupation', 'Height', 'Complexion', 'Eye_color', 'Hair_color'
    ])
    for url in tqdm(url_list):
        info_dict = single_url_crawler(url)
        black_soldiers = black_soldiers.append({
            'ID': info_dict.get('ID', ''),
            'soldier_name': info_dict.get('soldier_name', ''),
            'Age': info_dict.get('Age', ''),
            'Unit': info_dict.get('Unit', ''),
            'Enlisted_term_years': info_dict.get('Enlisted_term_years', ''),
            'Regiment': info_dict.get('Regiment', ''),
            'Company': info_dict.get('Company', ''),
            'Year': info_dict.get('Year', ''),
            'Year_Enlisted': info_dict.get('Year_Enlisted', ''),
            'Enlisted_date': info_dict.get('Enlisted_date', ''),
            'Enlisted_state': info_dict.get('Enlisted_state', ''),
            'Enlisted_county': info_dict.get('Enlisted_county', ''),
            'Enlisted_town': info_dict.get('Enlisted_town', ''),
            'State_born': info_dict.get('State_born', ''),
            'County_born': info_dict.get('County_born', ''),
            'Town_born': info_dict.get('Town_born', ''),
            'Occupation': info_dict.get('Occupation', ''),
            'Height': info_dict.get('Height', ''),
            'Complexion': info_dict.get('Complexion', ''),
            'Eye_color': info_dict.get('Eye_color', ''),
            'Hair_color': info_dict.get('Hair_color', '')
        }, ignore_index=True)
        
        # 实时写入
        black_soldiers.to_csv(r'C:\Users\lenovo\Desktop\黑人_crawler\black_soldiers.csv',index = False)
```

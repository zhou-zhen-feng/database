```python
import requests
from tqdm import tqdm
import time
import json
import pandas as pd
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv(r"D:\种族态度的经济驱动力研究\newspaper_sn.csv",encoding='ISO-8859-1')
```

```python
sns = list(df["newspaper_sn"])
```

```python
texts = []
for sn in tqdm(sns):
    url = f"https://chroniclingamerica.loc.gov/lccn/{sn}/"

    response = requests.get(url, verify=False)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        target_div = soup.find('div', {'class': 'about_issue_more row'})
        if target_div:
            text_detail = target_div.get_text()
            
    else:
        print(f"请求失败，状态码：{response.status_code}")
        text_detail = None
    
    texts.append(text_detail)
    
    time.sleep(1)
```

```python
len(texts)
```

```python
# 在 df 中新增一列
df['newspaper_introduction'] = texts
df
```

```python
import os
import numpy as np
import pandas as pd
import librosa
import librosa.display
import pylab as plt
import numpy as np
from tqdm.notebook import tqdm
from IPython.display import Audio
```

```python
# 获取包含的音频的文件路径，将音频加载为波形“y”并将采样率（单位：Hz）存储为“sr”
y, sr = librosa.load(r'C:\Users\Administrator\Desktop\运动类视频\audio\6883800380275215616.mp3')
```

```python
# 计算音频平均“节奏”
tempo, beat_frames = librosa.beat.beat_track(y=y, sr=sr)
tempo
```

```python
# 计算整个音频的平均响度
rms = librosa.feature.rms(y=y)
mean_rms = np.mean(rms)
mean_rms
```

```python
# 大规模运行
```

```python
dic = {}
folder = r'C:\Users\Administrator\Desktop\运动类视频\audio'
for audio in tqdm(os.listdir(folder)):
    try:
        
        audio_path = os.path.join(folder,audio)
        y, sr = librosa.load(audio_path)
        tempo, beat_frames = librosa.beat.beat_track(y=y, sr=sr)
        tempo_value = tempo.item()
        rms = librosa.feature.rms(y=y)
        mean_rms = np.mean(rms)

        id = audio.split('.')[0]
        dic[id] = [tempo_value,mean_rms]
        
    except Exception as e:
        print(f"出错了: {audio}, 错误信息: {e}")
        id = audio.split('.')[0]
        dic[id] = [None,None]
        
```
